```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
INF = 8
```

```python
def plot(
        signal, 
        title:str='', 
        y_range=(-1, 3), 
        figsize = (8, 3),
        x_label='n (Time Index)',
        y_label='x[n]',
        saveTo=None
    ):
    plt.figure(figsize=figsize)
    plt.xticks(np.arange(-INF, INF + 1, 1))
    
    y_range = (y_range[0], max(np.max(signal), y_range[1]) + 1)
    # set y range of 
    plt.ylim(*y_range)
    plt.stem(np.arange(-INF, INF + 1, 1), signal)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    if saveTo is not None:
        plt.savefig(saveTo)
    plt.show()
```

```python
signal = np.zeros(2*INF+1)
signal[INF] = 1
signal[INF+1] = .5
signal[INF-1] = 2
signal[INF + 2] = 1
signal[INF - 2] = .5
```

```python
plot(signal, 'OG x[n]')
```
